```python
unique_chars = []
for lyric in df["Lyrics"]:
    for char in str(lyric):  # Ensure the value is a string
        if char.isspace():
            continue  # Skip spaces or any whitespace
        if char not in unique_chars:
            unique_chars.append(char)

print(", ".join(unique_chars))
```

```python
class Preprocessing:
    def clean_text(text):
        # 1. Remove all characters except:
        #    - English letters (A-Z, a-z)
        #    - Arabic letters (Unicode range: \u0600-\u06FF)
        #    - Digits (0-9)
        #    - Whitespace (to preserve word separation)
        #   - Square brackets: [ and ]
        allowed_punctuation = "[]"  # Only square brackets are allowed
        pattern = r"[^A-Za-z0-9\u0600-\u06FF\s" + re.escape(allowed_punctuation) + "]"
        text = re.sub(pattern, '', text)

        # 2. Normalize Arabic letters by replacing variant forms with a standard form.
        text = re.sub(r"[إأٱآا]", "ا", text)
        text = re.sub(r"ى", "ي", text)
        text = re.sub(r"ؤ", "ء", text)
        text = re.sub(r"ئ", "ء", text)
        text = re.sub(r"ة", "ه", text)

        # 3. Remove Arabic diacritics using a regex pattern.
        # The following regex matches common diacritics and the tatwil (ـ) character.
        noise = re.compile(r""" ّ    | # Tashdid
                                 َ    | # Fatha
                                 ً    | # Tanwin Fath
                                 ُ    | # Damma
                                 ٌ    | # Tanwin Damm
                                 ِ    | # Kasra
                                 ٍ    | # Tanwin Kasr
                                 ْ    | # Sukun
                                 ـ     # Tatwil/Kashida
                              """, re.VERBOSE)
        text = re.sub(noise, '', text)

                #   - Square brackets: [ and ]
        allowed_punctuation = "[]"  # Only square brackets are allowed
        pattern = r"[^A-Za-z0-9\u0600-\u06FF\s" + re.escape(allowed_punctuation) + "]"

        # 4. Further remove any remaining diacritics (tashkeel) using pyarabic.
        text = araby.strip_tashkeel(text)

        return text
    def preprocess(sentence):
        return Preprocessing.clean_text(sentence)
```

```python
df["Lyrics"] = df["Lyrics"].apply(Preprocessing.preprocess)
```

```python
dataset_dict = {
    "Artist": df["Artist"].to_list(),
    "Lyrics": df["Lyrics"].tolist(),
    "Song_Title": df["Song Title"].to_list()
}
```

```python
from datasets import DatasetDict, Dataset
hf_dataset = Dataset.from_dict(dataset_dict)
```

```python
hf_dataset
```

```python
dataset_size = len(df)
train_size = int(0.8 * dataset_size)
val_size = int(0.1 * dataset_size)
test_size = dataset_size - train_size - val_size
```

```python
train_dataset = hf_dataset.select(range(train_size))
validation_dataset = hf_dataset.select(range(train_size, train_size + val_size))
test_dataset = hf_dataset.select(range(train_size + val_size, len(hf_dataset)))
print("Training set size:", train_size)
print("Validation set size:", val_size)
print("Testing set size:", test_size)
```

```python
dataset = DatasetDict({
    "train": train_dataset,
    "validation": validation_dataset,
    "test": test_dataset
})

for split_name, split_dataset in dataset.items():
    print(f"{split_name}: {split_dataset}")

features = dataset["train"].features
dataset = dataset.map(lambda example: {k: v for k, v in example.items() if k in features})
```
